```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

data = np.loadtxt('mlcc.txt')


x = data[:, 0]
y = data[:, 1]

xnew = np.linspace(x.min(), x.max(), 300)

gfg = make_interp_spline(x, y, k=2)

y_new = gfg(xnew)
plt.scatter(x, y, label = "Observed data points")  
plt.plot(xnew, y_new, color = "orange", label = "Trend line")


plt.ylabel("Capacitance (pF)")
plt.xlabel("Frequency (kHz)")
plt.legend()

plt.savefig("plot-mlcc.png", dpi = 1200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from scipy.optimize import curve_fit

data1 = np.loadtxt('5khz.txt')
data2 = np.loadtxt('15khz.txt')
data3 = np.loadtxt('25khz.txt')
data4 = np.loadtxt('35khz.txt')


x1 = data1[:, 0]
y1 = data1[:, 1]

x2 = data2[:, 0]
y2 = data2[:, 1]

x3 = data3[:, 0]
y3 = data3[:, 1]

x4 = data4[:, 0]
y4 = data4[:, 1]


xnew1 = np.linspace(x1.min(), x1.max(), 300)

gfg = make_interp_spline(x1, y1, k=2)

y_new1 = gfg(xnew1)




xnew2 = np.linspace(x2.min(), x2.max(), 300)

gfg = make_interp_spline(x2, y2, k=2)

y_new2 = gfg(xnew2)




xnew3 = np.linspace(x3.min(), x3.max(), 300)

gfg = make_interp_spline(x3, y3, k=2)

y_new3 = gfg(xnew3)




xnew4 = np.linspace(x4.min(), x4.max(), 300)

gfg = make_interp_spline(x4, y4, k=2)

y_new4 = gfg(xnew4)



plt.plot(xnew1, y_new1, label="5 kHz")
plt.plot(xnew2, y_new2, label="15 kHz")
plt.plot(xnew3, y_new3, label="25 kHz")
plt.plot(xnew4, y_new4, label="35 kHz")




plt.ylabel("Dielectric constant, $\epsilon$")
plt.xlabel("Temperature ($\degree C$)")
plt.legend()

plt.savefig("plot-curie.png", dpi = 1200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from scipy.optimize import curve_fit

data1 = np.loadtxt('5diffuse.txt')
data2 = np.loadtxt('15diffuse.txt')
data3 = np.loadtxt('25diffuse.txt')
data4 = np.loadtxt('35diffuse.txt')


x1 = data1[:, 0]
y1 = data1[:, 1]

x2 = data2[:, 0]
y2 = data2[:, 1]

x3 = data3[:, 0]
y3 = data3[:, 1]

x4 = data4[:, 0]
y4 = data4[:, 1]



plt.plot(x1, y1, "o")
plt.plot(x2, y2, "o")
plt.plot(x3, y3, "o")
plt.plot(x4, y4, "o")




m1, b1 = np.polyfit(x1, y1, 1)
plt.plot(x1, m1*x1 + b1, label = "5 kHz, slope = %0.2f" %m1)

print(m1, b1)





m2, b2 = np.polyfit(x2, y2, 1)
plt.plot(x2, m2*x2 + b2, label = "15 kHz, slope = %0.2f" %m2)

print(m2, b2)




m3, b3 = np.polyfit(x3, y3, 1)
plt.plot(x3, m3*x3 + b3, label = "25 kHz, slope = %0.2f" %m3)

print(m3, b3)




m4, b4 = np.polyfit(x4, y4, 1)
plt.plot(x4, m4*x4 + b4, label = "35 kHz, slope = %0.2f" %m4)

print(m4, b4)




plt.ylabel("ln ($1/\epsilon - 1/\epsilon_{max}$)")
plt.xlabel("ln($T-T_C$)")
plt.legend()

plt.savefig("plot-diffuse.png", dpi = 1200,bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('5diffuse.txt')
data2 = np.loadtxt('15diffuse.txt')
data3 = np.loadtxt('25diffuse.txt')
data4 = np.loadtxt('35diffuse.txt')


x1 = data1[:, 0]
y1 = data1[:, 1]

x2 = data2[:, 0]
y2 = data2[:, 1]

x3 = data3[:, 0]
y3 = data3[:, 1]

x4 = data4[:, 0]
y4 = data4[:, 1]

p1, V1 = np.polyfit(x1, y1, 1, cov=True)

print("m_1: {} +/- {}".format(p1[0], np.sqrt(V1[0][0])))
print("b_1: {} +/- {}".format(p1[1], np.sqrt(V1[1][1])))

p2, V2 = np.polyfit(x2, y2, 1, cov=True)

print("m_2: {} +/- {}".format(p2[0], np.sqrt(V2[0][0])))
print("b_2: {} +/- {}".format(p2[1], np.sqrt(V2[1][1])))

p3, V3 = np.polyfit(x3, y3, 1, cov=True)

print("m_3: {} +/- {}".format(p3[0], np.sqrt(V3[0][0])))
print("b_3: {} +/- {}".format(p3[1], np.sqrt(V3[1][1])))

p4, V4 = np.polyfit(x4, y4, 1, cov=True)

print("m_4: {} +/- {}".format(p4[0], np.sqrt(V4[0][0])))
print("b_4: {} +/- {}".format(p4[1], np.sqrt(V4[1][1])))
```
